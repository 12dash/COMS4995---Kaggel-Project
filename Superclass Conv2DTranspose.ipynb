```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential

import numpy as np
import pandas as pd

import utils 
```

```python
dir_ = "data_split"
data_dir = f"{dir_}/train/superclass"


image_size = 8
batch_size = 16

save_path = 'models/superclass/'
model_name = "superclass_Conv2DTranpose"
```

```python
train_ds, val_ds, num_classes = utils.get_ds(data_dir, save_path, 
                          image_size=image_size, 
                          batch_size=batch_size)
```

```python
AUTOTUNE = tf.data.AUTOTUNE

train_ds = train_ds.cache().shuffle(1000).prefetch(buffer_size=AUTOTUNE)
val_ds = val_ds.cache().prefetch(buffer_size=AUTOTUNE)
```

```python
model = Sequential([
  layers.Input(shape = (image_size,image_size,3)),
  layers.Conv2DTranspose(16, 2, padding='same', activation='relu'),    
  layers.Flatten(),
  layers.Dropout(0.6),
  layers.Dense(512,activation='relu'),
  layers.Dense(3)
])
```

```python
def scheduler(epoch, lr):
    lr = lr if epoch < 5 else 1e-4
    return lr

lr = tf.keras.callbacks.LearningRateScheduler(scheduler)
es = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=20)
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])

epochs = 100
history = model.fit(train_ds, validation_data=val_ds,
                    epochs=epochs, 
                    callbacks=[lr])
model.save(save_path+model_name)
```
