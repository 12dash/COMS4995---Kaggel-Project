```python
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
import tensorflow_addons as tfa

from tqdm.notebook import tqdm
import numpy as np
import pandas as pd

import utils 
```

```python
dir_ = "data_split"
data_dir = f"{dir_}/train/superclass"

image_size = 8
batch_size = 16

save_path = 'models/superclass/'
model_name = "superclass_skip_connection"
```

```python
train_ds, val_ds, num_class = utils.get_ds(data_dir, save_path,  image_size=image_size, 
                                batch_size=batch_size)
```

```python
AUTOTUNE = tf.data.AUTOTUNE

train_ds = train_ds.cache().shuffle(1000).prefetch(buffer_size=AUTOTUNE)
val_ds = val_ds.cache().prefetch(buffer_size=AUTOTUNE)
```

```python
inp = layers.Input(shape=(8,8,3))

x_1 = layers.Conv2DTranspose(16, 2, padding='same', activation='relu')(inp)
x_1 = layers.Conv2DTranspose(32, 2, padding='same', activation='relu')(x_1)
x_1 = layers.Conv2D(32, 2, padding='same', activation='relu')(x_1)
x_1 = layers.Conv2D(16, 2, padding='same', activation='relu')(x_1)

x = layers.Concatenate()([inp, x_1])

x_2 = layers.Conv2DTranspose(16, 2, padding='same', activation='relu')(x)
x_2 = layers.Conv2DTranspose(32, 2, padding='same', activation='relu')(x_2)
x_2 = layers.Conv2D(32, 2, padding='same', activation='relu')(x_2)
x_2 = layers.Conv2D(16, 2, padding='same', activation='relu')(x_2)

x = layers.Concatenate()([x, x_2])

x_3 = layers.Conv2DTranspose(16, 2, padding='same', activation='relu')(x)
x_3 = layers.Conv2DTranspose(32, 2, padding='same', activation='relu')(x_3)
x_3 = layers.Conv2D(32, 2, padding='same', activation='relu')(x_3)
x_3 = layers.Conv2D(16, 2, padding='same', activation='relu')(x_3)

x = layers.Concatenate()([x, x_3])

x_4 = layers.Conv2DTranspose(16, 2, padding='same', activation='relu')(x)
x_4 = layers.Conv2DTranspose(32, 2, padding='same', activation='relu')(x_4)
x_4 = layers.Conv2D(32, 2, padding='same', activation='relu')(x_4)
x_4 = layers.Conv2D(16, 2, padding='same', activation='relu')(x_4)

x = layers.Concatenate()([x, x_4])

x_5 = layers.Conv2DTranspose(16, 2, padding='same', activation='relu')(x)
x_5 = layers.Conv2DTranspose(32, 2, padding='same', activation='relu')(x_5)
x_5 = layers.Conv2D(32, 2, padding='same', activation='relu')(x_5)
x_5 = layers.Conv2D(16, 2, padding='same', activation='relu')(x_5)

x = layers.Concatenate()([x, x_5])

x_6 = layers.Conv2DTranspose(16, 2, padding='same', activation='relu')(x)
x_6 = layers.Conv2DTranspose(32, 2, padding='same', activation='relu')(x_6)
x_6 = layers.Conv2D(32, 2, padding='same', activation='relu')(x_6)
x_6 = layers.Conv2D(16, 2, padding='same', activation='relu')(x_6)

x = layers.Flatten()(x_6)
out = layers.Dense(3)(x)

model = tf.keras.Model(inputs=inp, outputs=out, name="skip_connection")
```

```python
def scheduler(epoch, lr):
    lr = lr if epoch < 5 else 1e-5 
    return lr

lr = tf.keras.callbacks.LearningRateScheduler(scheduler)
es = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=20)
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4),
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])

epochs = 200
history = model.fit(train_ds, validation_data=val_ds, 
                    epochs=epochs, callbacks=[lr])
model.save(save_path+model_name)
```
