```python
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
import tensorflow_addons as tfa

from tqdm.notebook import tqdm
import numpy as np
import pandas as pd

import utils 
```

```python
dir_ = "data_split"
data_dir = f"{dir_}/train/superclass"


image_size = 8
batch_size = 16

save_path = 'models/superclass/'
model_name = "superclass_contrastive"
```

```python
train_ds, val_ds, num_class = utils.get_ds(data_dir, save_path,  image_size=image_size, 
                                batch_size=batch_size)
```

```python
AUTOTUNE = tf.data.AUTOTUNE

train_ds = train_ds.cache().shuffle(1000).prefetch(buffer_size=AUTOTUNE)
val_ds = val_ds.cache().prefetch(buffer_size=AUTOTUNE)
```

```python
dropout_rate = 0.5
learning_rate = 1e-4
```

```python
def create_encoder():
    model_feature = Sequential([
      layers.Conv2DTranspose(16, 2, padding='same', activation='relu'),
      layers.Flatten()
    ])
    
    inputs = keras.Input(shape=(image_size, image_size, 3))
    outputs = model_feature(inputs)
    model = keras.Model(inputs=inputs, outputs=outputs)
    
    return model
```

```python
class SupervisedContrastiveLoss(keras.losses.Loss):
    def __init__(self, temperature=1, name=None):
        super(SupervisedContrastiveLoss, self).__init__(name=name)
        self.temperature = temperature

    def __call__(self, labels, feature_vectors, sample_weight=None):
        feature_vectors_normalized = tf.math.l2_normalize(feature_vectors, axis=1)
        logits = tf.divide(tf.matmul(feature_vectors_normalized, tf.transpose(feature_vectors_normalized)),
                           self.temperature)
        return tfa.losses.npairs_loss(tf.squeeze(labels), logits)


def add_projection_head(encoder):
    inputs = keras.Input(shape=(image_size, image_size, 3))
    features = encoder(inputs)
    outputs = layers.Dense(256, activation="relu")(features)
    model = keras.Model(inputs=inputs, outputs=outputs, name="projection")
    return model
```

```python
def create_classifier(encoder, trainable=True):
    for layer in encoder.layers:
        layer.trainable = trainable
        
    inputs = keras.Input(shape=(image_size, image_size, 3))
    features = encoder(inputs)
    features = layers.Dropout(0.5)(features)
    features = layers.Dense(512, activation="relu")(features)
    features = layers.Dropout(0.4)(features)
    features = layers.Dense(256, activation="relu")(features)
    outputs = layers.Dense(3, activation="softmax")(features)

    model = keras.Model(inputs=inputs, outputs=outputs)
    model.compile(optimizer=keras.optimizers.Adam(learning_rate), 
                  loss=keras.losses.SparseCategoricalCrossentropy(),
                  metrics=['accuracy'])
    return model
```

```python
encoder = create_encoder()

encoder_with_projection_head = add_projection_head(encoder)
encoder_with_projection_head.compile(optimizer=keras.optimizers.Adam(1e-4),
                                     loss=SupervisedContrastiveLoss(0.15))

encoder_with_projection_head.summary()
```

```python
history = encoder_with_projection_head.fit(train_ds, validation_data=val_ds, epochs=100)
```

```python
classifier = create_classifier(encoder, trainable=False)

history = classifier.fit(train_ds, validation_data=val_ds,epochs=200)
accuracy = classifier.evaluate(val_ds)
```

```python
utils.plot(history)
```
